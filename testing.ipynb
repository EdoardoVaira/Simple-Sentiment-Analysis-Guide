```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

```python
import emoji
import re
import contractions
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# Ensure you have these resources downloaded
nltk.download('punkt')
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def text_cleaner(text):
    # Convert text to lowercase
    text = text.lower()
    # Replace emojis with text description
    text = emoji.demojize(text, delimiters=("", " "))
    # Expand contractions
    text = contractions.fix(text)
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Replace punctuation with a space (and remove extra spaces)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()  # Normalize white spaces
    # Normalize excessive letter repetitions
    text = re.sub(r'(.)\1+', r'\1\1', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Tokenize text
    tokens = word_tokenize(text)
    # Lemmatize
    cleaned_text = [lemmatizer.lemmatize(word) for word in tokens]
    # Join words back to form the cleaned text
    return ' '.join(cleaned_text)
```

```python
df['Content_cleaned'] = df['Content'].apply(text_cleaner)
```
